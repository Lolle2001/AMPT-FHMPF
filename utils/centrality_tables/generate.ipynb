```python
import numpy as np
import pandas as pd
import glob

def read_latex_table(filename):
    first_index = 0
    last_index = 0

    with open(filename, 'r') as f:
        lines = f.readlines()
        for i in range(len(lines)):
            sline = lines[i].split()
            if sline[0] == 'Centrality':
                first_index = i + 1
            if sline[0] ==  r'\end{tabular}':
                last_index = i
            
        nlines = lines[first_index:last_index]
        rangeflt = []
        rangestr = []
        for nline in nlines:
            sline = nline.split('&')
            
            flt = [float(sline[1]), float(sline[2])]

            str = sline[0].replace('\\%','').split('--')
            str = [float(str[0]), float(str[1])]
            rangeflt.append(flt)
            rangestr.append(str)
            
        rangeflt = np.array(rangeflt)
        rangestr = np.array(rangestr)

        sname = filename.split("_")
        energy = sname[1].split(".")[0].replace('GeV','')
        ssname = sname[0].split("-")
        projectile = ssname[0]
        target = ssname[1]
        n = len(rangeflt)
        rangeproj = [projectile for i in range(n)]
        rangetarg = [target for i in range(n)]
        rangeenergy = [energy for i in range(n)]
        df = pd.DataFrame(columns = ['Target', 'Projectile', 'sqrt(sNN)', 'cenmin', 'cenmax', 'bmin', 'bmax'])
        
        df['Target'] = rangetarg
        df['Projectile'] = rangeproj
        df['sqrt(sNN)'] = rangeenergy
        df['cenmin'] = rangestr[:,0]
        df['cenmax'] = rangestr[:,1]
        df['bmin'] = rangeflt[:,0]
        df['bmax'] = rangeflt[:,1]
    f.close()
    with open(filename.replace('.tab', '.arr'), 'w') as f:
        f.write(f"# PROJECTILE: {projectile}\n")
        f.write(f"# TARGET    : {target}\n")
        f.write(f"# ENERGY    : {energy} GeV\n")
        f.write(f"# CENTRALITY(%)\tIMPACT PARAMETER(fm) \n")
        for i in range(n - 1):
            f.write(f"{rangestr[i][0]:.0f}\t{rangeflt[i][0]:.2f}\n")
        f.write(f"{rangestr[n-1][1]:.0f}\t{rangeflt[n-1][1]:.2f}\n")
    f.close()
    return df



```

```python





files = glob.glob("*.tab")

df = pd.DataFrame()

dflist = []
print(files)
for file in files:
    dflist.append(read_latex_table(file))
    print(file)

df = pd.concat(dflist)

```
